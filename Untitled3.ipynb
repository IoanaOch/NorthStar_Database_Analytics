```r
# Install required packages
install.packages(c("sqldf","dplyr","ggplot2","tidyverse"))

# Load libraries
library(sqldf)
library(dplyr)
library(ggplot2)
library(tidyverse)

# Read datasets
customers <- read.csv("customers.csv")
orders <- read.csv("orders.csv")
deliveries <- read.csv("deliveries.csv")
drivers <- read.csv("drivers.csv")
vehicles <- read.csv("vehicles.csv")
hubs <- read.csv("hubs.csv")
complaints <- read.csv("complaints.csv")
incidents <- read.csv("incidents.csv")
app_events <- read.csv("app_events.csv")
data_dictionary <- read.csv("data_dictionary.csv")

cat("All datasets loaded successfully!")
```

```r
# Display first rows of datasets

head(customers)
head(orders)
head(deliveries)

# Show structure of datasets

str(customers)
str(orders)
str(deliveries)
```

```r
# Query 1
# Average order value by service type

query1 <- sqldf("

SELECT
service_type,
ROUND(AVG(order_value),2) AS average_order_value

FROM orders

GROUP BY service_type

ORDER BY average_order_value DESC

")

print(query1)
```

```r
# Query 2
# Count deliveries by delivery status

query2 <- sqldf("

SELECT
delivery_status,
COUNT(*) AS total_deliveries

FROM deliveries

GROUP BY delivery_status

ORDER BY total_deliveries DESC

")

print(query2)
```

```r
# Query 3
# Average manual route overrides by hub

query3 <- sqldf("

SELECT
hub_id,
ROUND(AVG(manual_route_override_count),2) AS average_route_overrides

FROM deliveries

GROUP BY hub_id

ORDER BY average_route_overrides DESC

")

print(query3)
```

```r
# Query 4
# Average customer rating by delivery status

query4 <- sqldf("

SELECT
delivery_status,
ROUND(AVG(customer_rating_post_delivery),2) AS average_customer_rating

FROM deliveries

WHERE customer_rating_post_delivery IS NOT NULL

GROUP BY delivery_status

ORDER BY average_customer_rating DESC

")

print(query4)
```

```r
# Query 5
# Average fuel or charge cost by delivery status

query5 <- sqldf("

SELECT
delivery_status,
ROUND(AVG(fuel_or_charge_cost),2) AS average_cost

FROM deliveries

GROUP BY delivery_status

ORDER BY average_cost DESC

")

print(query5)
```

```r
# Graph 1
# Distribution of delivery status

ggplot(deliveries, aes(x=delivery_status)) +
geom_bar() +
labs(
title="Distribution of Delivery Status",
x="Delivery Status",
y="Number of Deliveries"
)
```

```r
# Graph 2
# Customer rating by delivery status

ggplot(deliveries,
aes(x=delivery_status,
y=customer_rating_post_delivery)) +
geom_boxplot() +
labs(
title="Customer Rating by Delivery Status",
x="Delivery Status",
y="Customer Rating"
)
```
